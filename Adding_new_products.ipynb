## **Adding new items to inventory**

```python
from datetime import datetime ,date


id =str(input("Enter product id item: "))
name = str(input("Enter name: "))
price = int(input("Enter price: "))
qn = int(input("Enter quantity: "))
manf_str = str(input('Enter manufacturing date(yyyy-mm-dd): '))
manf_dt = datetime.strptime(manf_str, "%Y-%m-%d").date()
exp_str = str(input('Enter expiry date(yyyy-mm-dd): '))
exp_dt = datetime.strptime(exp_str, "%Y-%m-%d").date()

if id in record.keys():
  print("Key already exists!!!")
  print("Edit the previous record")
  ip = str(input("To edit type - Yes or No: "))
  if ip == 'Yes':
    #for editing price
    price = price + record[id]['pr']
    #for editing quantity
    qn = record[id]['qn']+qn
    record[id] = {'name': name,'pr': price,'qn': qn,'manf_date':manf_dt.isoformat(),'exp_date':exp_dt.isoformat()}
  elif ip=='No':
    pass
  else:
    print("Invalid Input")
else:
  record[id] = {'name': name,'pr': price,'qn': qn,'manf_date':manf_dt.isoformat(),'exp_date':exp_dt.isoformat()}


js = json.dumps(record,default=str)

fd= open("record.json",'w')
fd.write(js)
fd.close()
```
